```python
# Dependencies
import pandas as pd
```

```python
# load CSV
csv_path = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas
dataunclean_df = pd.read_csv(csv_path, low_memory=False)
dataunclean_df.head()
```

```python
# Inspect all columns
for col in dataunclean_df.columns: 
    print(col) 
```

```python
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111


dataclean_df = dataunclean_df.iloc[:,[0, 1, 2, 3, 4, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
dataclean_df
```

```python
# Change "0" to "No" and "1" to "Yes" in response columns
dataclean_df = dataclean_df.replace({0.0 : "No", 1.0 : "Yes"})
dataclean_df.head()
```

```python
# Calculate total number of respondents in survey
number_respondents=len(dataclean_df)
print(number_respondents)
```

```python
# Extract rows corresponding only to people who attended a bootcamp
Attend_bootcamp = dataclean_df.loc[dataclean_df["AttendedBootcamp"] == "Yes"]
Attend_bootcamp.count()
```

```python
# Calculate average age of attendees
mean_age = dataclean_df["Age"].mean()
print(mean_age)

# Calculate how many people attended a bootcamp
Boot_camp_attendees = Attend_bootcamp["AttendedBootcamp"].count()
print(Boot_camp_attendees)

#Calculate how many attendees hold degrees
Attendees_hold_degrees = Attend_bootcamp["SchoolDegree"].count()
print(Attendees_hold_degrees)

# Count number of attendees who self-identify as male; female; or are of non-binary gender identification
Gender_count = Attend_bootcamp['Gender'].count()
Male_count = Attend_bootcamp['Gender'].value_counts()["male"]
Female_count = Attend_bootcamp['Gender'].value_counts()["female"]
NonBinary_count = Gender_count-(Male_count + Female_count)
print(Gender_count)
print(Male_count)
print(Female_count)
print(NonBinary_count)

# Calculate percentage of respondents who attended a bootcamp
Percent_respondents = (Boot_camp_attendees/number_respondents)*100
print(Percent_respondents)

# Calculate percentage of respondents belonging to each gender
Male_Percent = (Male_count/number_respondents) * 100
Female_Percent = (Female_count/number_respondents) * 100
NonBinary_Percent = (NonBinary_count/number_respondents) * 100

# Calculate percentage of attendees with a school degree
School_degree_percent = (Attendees_hold_degrees/Boot_camp_attendees)*100

# Calculate average post-bootcamp salary
Avg_salary = Attend_bootcamp["BootcampPostSalary"].mean()
print(Avg_salary)
```

```python
# Create a new table consolidating above calculations
new_df = pd.DataFrame([
    {"Average Salary": Avg_salary}, {"Percentage of School degree": School_degree_percent}, {"Male percent of respondents": Male_Percent}, {"Female percent of respondents": Female_Percent}, 
       {"NonBinary percent of respondents": NonBinary_Percent}, {"Number of people attending Bootcamp": Boot_camp_attendees}, 
    {"The number of attendees that hold degrees": Attendees_hold_degrees}, {"Average Age of Attendees": mean_age}])

# Improve formatting before outputting spreadsheet

pd.options.display.float_format = '${:,.2f}'.format
```

```python
# Export to Excel
new_df.to_excel("output/DataCleanUp.xls", sheet_name='Sheet_name_1',index=False, header=True)
```
